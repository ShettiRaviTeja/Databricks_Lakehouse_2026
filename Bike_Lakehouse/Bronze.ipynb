# Reading and Writing into Tables

```python
from bronze_config import INGESTION_CONFIG
for config in INGESTION_CONFIG:
    df = (
        spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(config["path"])
    )
    df.write.mode("overwrite").format("delta").saveAsTable(f"bronze.{config["table"]}")
```
